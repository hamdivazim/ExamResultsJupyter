```python
# Import Modules
import numpy as np
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

```python
#  Create Data Frame
results = pd.read_csv('ExamResults.csv')
```

```python
# Show Table
results
```

```python
# Add a column for Pass/Fail
results['Pass/Fail'] = pd.Index(results['Overall Marks'] > (85+results['Grade']-6))
results.head()
```

```python
# Plot Bar Chart
passed = results['Pass/Fail'].value_counts()
pass_index = passed.index
passed_values = passed.values

plt.figure(figsize=(20, 8))
plt.xticks(rotation=0)
plt.title('Pass Rate', fontsize=18)
plt.ylabel('Amount of Children', fontsize=16)
pass_plot = sb.barplot(x=pass_index, y=passed_values, palette='Blues', alpha=0.6)
plt.show()
```

```python
# Plot Bar Chart 2
subject = results['Exam Subject'].value_counts()
subject_index = subject.index
subject_values = subject.values

plt.figure(figsize=(20, 8))
plt.xticks(rotation=0)
plt.title('Subjects Taken', fontsize=18)
plt.xlabel('Subject', fontsize=16)
plt.ylabel('Amount of Children', fontsize=16)
pass_plot = sb.barplot(x=subject_index, y=subject_values, palette='Greens', alpha=0.6)
plt.show()
```

```python
# Plot Bar Chart 3
overall = results['Overall Marks'].value_counts()
overall_index = overall.index
overall_values = overall.values

plt.figure(figsize=(20, 8))
plt.xticks(rotation=0)
plt.title('Overall Marks', fontsize=18)
plt.xlabel('Marks', fontsize=16)
plt.ylabel('Amount of Children', fontsize=16)
pass_plot = sb.barplot(x=overall_index, y=overall_values, palette='viridis', alpha=0.6)
plt.show()
```

```python
# Add a column for Grading
grading = ['F', 'C', 'B', 'A', 'A*']
results['Grading'] = pd.Index([grading[int(i >= 90) + int(i >= 80) + int(i >= 70) + int(i >= 60)] for i in results['Overall Marks']])
results.head()
```

```python
# Plot Bar Chart 4
overall = results['Grading'].value_counts()
overall_index = overall.index
overall_values = overall.values

plt.figure(figsize=(20, 8))
plt.xticks(rotation=0)
plt.title('Overall Grading', fontsize=18)
plt.xlabel('Grade', fontsize=16)
plt.ylabel('Amount of Children', fontsize=16)
pass_plot = sb.barplot(x=overall_index, y=overall_values, palette='gist_ncar_r', alpha=0.6, order=['A*', 'A', 'B', 'C', 'F'])
plt.show()
```

```python
# Create a Line Graph 'catplot'
sb.catplot(x='First Name', y='Overall Marks', data=results, kind='point', height=8, aspect=3)
```
